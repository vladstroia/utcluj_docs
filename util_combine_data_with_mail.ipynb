```python
import pandas as pd

df = pd.read_csv('tabel_materii.csv', dtype=str)
df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df_mail = pd.read_csv('tabel_adrese_gasite.csv', dtype=str, sep = '---')
# we have to create the context object for each row in the csv file

```

```python
# df_mail['nume_curs']['Software pentru sisteme mecatronice']
numele_cursului = '    Bazele cercetarii experimentale     '

# df_mail.loc[df_mail['nume_curs'] == numele_cursului]['email'].iloc[0]
if df_mail.loc[df_mail['nume_curs'] == numele_cursului.strip()]['email'].empty:
    print('nu exista cursul')
else:
    print('exista cursul')
```

```python
date_curs = []
i = 0
while i < 1898:
    materie = {}
    while True:
        rand = df.loc[[i]]
        for item in rand:
            if item == 'formapred':
                materie[rand['formapred'].values[0].strip()] = rand["numarore"].values[0]
            else:
                materie[item] = rand[item].values[0]
        if df['nrcrt'][i + 1] != df['nrcrt'][i]:
            break
        else:
            i+=1
    i +=1
    numele_cursului = materie['nume_disciplina']
    print(numele_cursului)
    if not df_mail.loc[df_mail['nume_curs'] == numele_cursului.strip()]['email'].empty:
        materie['email'] = df_mail.loc[df_mail['nume_curs'] == numele_cursului.strip()]['email'].iloc[0]
    else:
        materie['email'] = ''
    date_curs.append(materie)

new_df = pd.DataFrame(date_curs)
new_df
```

```python
new_df.to_csv('tabel_materii_si_email.csv')
```
